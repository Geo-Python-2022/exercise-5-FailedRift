The output should contain mean, max, and min Celsius temperatures for each day (for example, one mean temperature value for the 1st of May and so on).

````python
import pandas as pd
````

````python
#read csvs
kumpula = pd.read_csv("Kumpula_temps_May_Aug_2017.csv")
rovaniemi = pd.read_csv("Rovaniemi_temps_May_Aug_2017.csv")
````

````python
kumpula.head()
````

````python
#Change column name
kumpula["Unnamed: 0"] = kumpula["USAF"]
rovaniemi["Unnamed: 0"] = rovaniemi["USAF"]
````

````python
#Drop unnamed axis
kumpula.drop("Unnamed: 0", axis=1, inplace=True)
rovaniemi.drop("Unnamed: 0", axis=1, inplace=True)
````

````python
kumpula
````

````python
rovaniemi
````

````python
#Aggregating on date
kgrp = kumpula.groupby("YR--MODAHRMN")
rgrp = rovaniemi.groupby("YR--MODAHRMN")
````

````python
#Mean celsius temperatures by date
kgrp_mean = (kgrp["Celsius"].mean())
rgrp_mean = (rgrp["Celsius"].mean())
````

````python
#Maximum Celsius Temperatures by date
kgrp_min = (kgrp["Celsius"].min())
rgrp_min = (rgrp["Celsius"].min())
````

````python
#Minimum Celsius Temperatures by date
kgrp_max = (kgrp["Celsius"].max())
rgrp_max = (rgrp["Celsius"].max())
````

````python
kgrp_mean.index
````

````python
rgrp_mean.index
````

````python
#import reduce (will learn what it actually does)
from functools import reduce

# Dataframes to merge
dfs_k = [kgrp_mean, kgrp_min, kgrp_max]
dfs_r = [rgrp_mean, rgrp_min, rgrp_max]

#Merge function. Will learn this. Copied from stackoverflow
kgrp_merged = reduce(lambda  left,right: pd.merge(left,right,on=['YR--MODAHRMN'],
                                            how='outer'), dfs_k)

rgrp_merged = reduce(lambda  left,right: pd.merge(left,right,on=['YR--MODAHRMN'],
                                            how='outer'), dfs_r)
````

````python
kgrp_merged
````

````python
rgrp_merged
````

````python
#Changed column names to MEAN, MIN, MAX
kgrp_merged.columns = ["MEAN", "MIN", "MAX"]
rgrp_merged.columns = ["MEAN", "MIN", "MAX"]
````
